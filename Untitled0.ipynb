```
import numpy as np

#criar e preencher o mundo de Wumpus
def criar_mundo():
    cols, rows = 20,20
    total_pit = 2

    wumpus_world = np.zeros((rows, cols), dtype=int)
    wumpus_world[0, 0] = 5

    remaining_positions = rows * cols - 1
    possible_positions = np.arange(1, rows * cols)
    pit_random_positions = np.random.choice(possible_positions, total_pit, replace=False)

    for pos in pit_random_positions:
        row = pos // cols
        col = pos % cols
        wumpus_world[row, col] = 3

    remaining_pos_wumps_gold = np.setdiff1d(possible_positions, pit_random_positions)
    np.random.shuffle(remaining_pos_wumps_gold)
    pos_random_wumpus_gold = remaining_pos_wumps_gold[:2]

    if 1 in pos_random_wumpus_gold:
        pos_random_wumpus_gold = [pos for pos in pos_random_wumpus_gold if pos != 0]
        pos_random_wumpus_gold = [0] + pos_random_wumpus_gold
    else:
        pos_random_wumpus_gold[0] = 0

    for pos, content in zip(pos_random_wumpus_gold, [1, 2]):
        row = pos // cols
        col = pos % cols
        wumpus_world[row, col] = content

    return wumpus_world
```

```
# Funções do jogo
def obter_vizinhos_posicoes(matriz, i, j):
    linhas = len(matriz)
    colunas = len(matriz[0])
    vizinhos_posicoes = []

    for x in range(max(0, i - 1), min(i + 2, linhas)):
        for y in range(max(0, j - 1), min(j + 2, colunas)):
            if (x, y) != (i, j):
                vizinhos_posicoes.append((x, y))

    return vizinhos_posicoes

def atualizar_matriz(matriz):
    linhas = len(matriz)
    colunas = len(matriz[0])
    nova_matriz = [[0 for _ in range(colunas)] for _ in range(linhas)]

    for i in range(linhas):
        for j in range(colunas):
            vizinhos = obter_vizinhos_posicoes(matriz, i, j)
            nova_matriz[i][j] = sum(matriz[x][y] for x, y in vizinhos)

    return nova_matriz

def agent_random_mov(wumpus_world, current_pos):
    directions = ['up', 'down', 'right', 'left']
    movement = np.random.choice(directions)
    new_pos = agent_mov(wumpus_world, current_pos, movement)
    return new_pos, movement

def agent_mov(wumpus_world, current_pos, action):
    rows, cols = wumpus_world.shape
    new_pos = list(current_pos)

    if action == 'up' and current_pos[0] > 0:
        new_pos[0] -= 1
    elif action == 'down' and current_pos[0] < rows - 1:
        new_pos[0] += 1
    elif action == 'left' and current_pos[1] > 0:
        new_pos[1] -= 1
    elif action == 'right' and current_pos[1] < cols - 1:
        new_pos[1] += 1

    return tuple(new_pos)

def detect_sensations(wumpus_world, current_pos):
    sensations = []
    rows, cols = wumpus_world.shape

    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue
            new_row = current_pos[0] + i
            new_col = current_pos[1] + j
            if 0 <= new_row < rows and 0 <= new_col < cols:
                sens = wumpus_world[new_row, new_col]
                if sens == 2:
                    sensations.append(7)
                elif sens == 3:
                    sensations.append(9)
                elif sens == 1:
                    sensations.append(10)

    return sensations

def game_over(wumpus_world, current_pos):
    mov = wumpus_world[current_pos]
    if mov == 2:
        return True, -1000
    elif mov == 3:
        return True, -1000
    return False, 0


    # Exibe o relatório
    for result in results:
        print(result)

```

```
# Função principal para executar o jogo
results = []
def jogar_mundo_wumpus():
    wumpus_world = criar_mundo()
    print("Mundo de Wumpus:")
    print(wumpus_world)

    score = 0
    current_pos = (0, 0)
    movements = []
    found_gold = False

    while True:
        score -= 1

        new_pos, movement = agent_random_mov(wumpus_world, current_pos)
        print(f"Movimento: {movement}, Nova Posição: {new_pos}")

        is_game_over, penalty = game_over(wumpus_world, new_pos)
        if is_game_over:
            score += penalty
            print(f"Pontuação: {score}")
            print(f"Movimento: {movements}, Nova posição: {current_pos}")
            print("### GAME OVER ###")
            break

        current_pos = new_pos
        movements.append(movement)

        sensations = detect_sensations(wumpus_world, current_pos)
        if 10 in sensations:
            print(f"Movimento: {movements}, Nova posição: {current_pos}, Sensações: {sensations}")
            print("### YOU FOUND THE GOLD ###")
            score += 1000
            found_gold = True
            break

        print(f"Nova Posição: {current_pos}, Sensações: {sensations}")
        print(wumpus_world)

    if found_gold:
        print("Retornando à posição inicial (0,0)...")
        score += 1000
        current_pos = (0, 0)
        print(f"Agente retornou para a posição inicial: {current_pos}")

    return score

# Executa várias vezes e gera o relatório final
def gerar_relatorio(execucoes):
    final_scores = []
    for exec_ag in range(execucoes):
        print(f"\nExecução {exec_ag + 1}:")
        score = jogar_mundo_wumpus()
        final_scores.append(score)


    results.append("\nResultados Finais:")
    results.append(f"Execuções: {execucoes}")
    results.append(f"Pontuação Final por Execução: {final_scores}")
    results.append(f"Pontuação Média: {np.mean(final_scores):.2f}")
```

```
# Executa 10 simulações e gera o relatório final
gerar_relatorio(20)
```
